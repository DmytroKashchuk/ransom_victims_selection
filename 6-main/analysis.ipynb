## Employees vs Group Correlation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data/data_tech_swdb.csv")

# bins = [0, 10, 50, 250, 1000, float("inf")]
bins = [0, 10, 50, float("inf")]  # --- IGNORE ---
labels = ["Small", "Medium", "Enterprise"]
df["size_class"] = pd.cut(df["account_employees"], bins=bins, labels=labels, right=False)

top15 = df["group_name"].value_counts().head(80).index
df_top = df[df["group_name"].isin(top15)]

ct = pd.crosstab(df_top["group_name"], df_top["size_class"])
ct_pct = ct.div(ct.sum(axis=1), axis=0).mul(100).round(1)
ct_pct.index = [f"{g} (n={ct.loc[g].sum()})" for g in ct_pct.index]

plt.figure(figsize=(10, 16))
sns.heatmap(ct_pct, annot=True, fmt=".1f", cmap="YlOrRd", cbar_kws={"label": "% vittime"})
plt.title("Preferenza dimensione azienda per gruppo ransomware")
plt.xlabel("Dimensione azienda")
plt.ylabel("Gruppo")
plt.tight_layout()
plt.savefig("heatmap_group_size.png", dpi=150)
plt.show()
```

# Revenue vs Group Correlation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data/data_tech_swdb.csv")

bins = [0, 1e6, 10e6, 50e6, 250e6, float("inf")]
labels = ["<1M", "1M-10M", "10M-50M", "50M-250M", "250M+"]
df["revenue_class"] = pd.cut(df["account_revenue_usd"], bins=bins, labels=labels, right=False)

top15 = df["group_name"].value_counts().head(80).index
df_top = df[df["group_name"].isin(top15)]

ct = pd.crosstab(df_top["group_name"], df_top["revenue_class"])
ct_pct = ct.div(ct.sum(axis=1), axis=0).mul(100).round(1)
ct_pct.index = [f"{g} (n={ct.loc[g].sum()})" for g in ct_pct.index]

plt.figure(figsize=(10, 16))
sns.heatmap(ct_pct, annot=True, fmt=".1f", cmap="YlOrRd", cbar_kws={"label": "% vittime"})
plt.title("Preferenza revenue azienda per gruppo ransomware")
plt.xlabel("Revenue")
plt.ylabel("Gruppo")
plt.tight_layout()
plt.savefig("heatmap_group_revenue.png", dpi=150)
plt.show()
```

# Spendings

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data/data_tech_swdb.csv")

top80 = df["group_name"].value_counts().head(80).index
df_top = df[df["group_name"].isin(top80)]

spending_cols = {
    "account_hw_spending": ("HW Spending", [0, 5e3, 25e3, 100e3, 500e3, float("inf")], ["<5K", "5K-25K", "25K-100K", "100K-500K", "500K+"]),
    "account_sw_spending": ("SW Spending", [0, 5e3, 25e3, 100e3, 500e3, float("inf")], ["<5K", "5K-25K", "25K-100K", "100K-500K", "500K+"]),
    "account_ict_spending": ("ICT Spending", [0, 25e3, 100e3, 500e3, 2e6, float("inf")], ["<25K", "25K-100K", "100K-500K", "500K-2M", "2M+"]),
    "account_it_services_spending": ("IT Services Spending", [0, 10e3, 50e3, 200e3, 1e6, float("inf")], ["<10K", "10K-50K", "50K-200K", "200K-1M", "1M+"]),
}

for col, (title, bins, labels) in spending_cols.items():
    df_top[f"{col}_class"] = pd.cut(df_top[col], bins=bins, labels=labels, right=False)
    ct = pd.crosstab(df_top["group_name"], df_top[f"{col}_class"])
    ct_pct = ct.div(ct.sum(axis=1), axis=0).mul(100).round(1)
    ct_pct.index = [f"{g} (n={ct.loc[g].sum()})" for g in ct_pct.index]

    plt.figure(figsize=(10, 16))
    sns.heatmap(ct_pct, annot=True, fmt=".1f", cmap="YlOrRd", annot_kws={"size": 7}, cbar_kws={"label": "% vittime"})
    plt.title(f"Preferenza {title} per gruppo ransomware")
    plt.xlabel(title)
    plt.ylabel("Gruppo")
    plt.tight_layout()
    plt.savefig(f"heatmap_group_{col}.png", dpi=150)
    plt.close()
    print(f"Saved heatmap_group_{col}.png")
```
