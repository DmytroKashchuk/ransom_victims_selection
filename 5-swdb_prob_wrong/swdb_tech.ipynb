```python
import pandas as pd
import json

df = pd.read_csv("data/swdb_prodcat.csv")

# Parse JSON column
records = []
for _, row in df.iterrows():
    techs = json.loads(row["technolgies"])
    for t in techs:
        records.append({"category": t["category"], "product": t["product"]})

tech_df = pd.DataFrame(records)

# Group by category: unique technologies and count
result = (
    tech_df.groupby("category")["product"]
    .apply(lambda x: x.unique())
    .reset_index()
)
result["n_tech"] = result["product"].apply(len)
result["technologies"] = result["product"].apply(lambda x: "; ".join(sorted(x)))
result = result[["category", "n_tech", "technologies"]].sort_values("n_tech", ascending=False)

result.to_csv("data/categories_technologies.csv", index=False)
print(result.to_string(index=False))
```

```python
import pandas as pd
import json
import numpy as np
from scipy.stats import fisher_exact

df = pd.read_csv("data/swdb_prodcat.csv")

relevant_cats = [
    "Application Development Tools", "Database Management", "Anti-Virus",
    "IT Infrastructure & Ops Management", "Server", "Router",
    "Collaborative/Integrated Software", "Operating System", "Network Attached Storage",
    "Enterprise Resource Planning", "Software as a Service", "Infrastructure as a Service",
    "E-Commerce Platform/Software", "CRM", "Document Management",
    "Network Management Software", "Web Hosting", "Other IT Development/Infrastructure",
    "Web Content Management Systems", "Security Information & Event Management (SIEM)",
    "Storage Management", "Voice over IP", "Email Software",
    "Frameworks & Libraries", "Server Virtualization", "Big Data Processing",
    "Platform as a Service", "Enterprise Management", "Testing & QA",
    "Data Integration", "Printer", "Software Configuration Management",
    "Help Desk Management", "Domain Name Services", "Other Security Software",
    "Content Delivery Networks", "Backup & Recovery",
    "Identity & Access Management (IAM)", "Unified Communications Services",
    "Governance Risk & Compliance", "Endpoint Security", "LAN Switch",
    "Website Builders", "IT Service Management (ITSM)", "Other Web Tools & Plugins",
    "Programming Languages", "Disaster Recovery", "Firewall Software",
    "Enterprise Mobility Management", "Storage Area Network",
    "Business Process Management", "Application Server Software",
    "Data Science & Machine Learning", "Web Services", "LAN Operating System",
    "Desktop Virtualization", "Advanced Threat Protection", "Virtual Private Network",
    "Server Management Tools", "IBM/PCM Mainframe", "Video Conferencing",
    "Desktop Management Tools", "Phone System", "Groupware Software",
    "Containerization", "Enterprise Application Integration", "Data Management",
    "Email Security", "Program Testing/Debugging", "Storage as a Service",
    "CAPTCHA", "Data Loss Prevention Software", "Portal Management"
]

# Parse and filter relevant techs per site
records = []
for _, row in df.iterrows():
    techs = json.loads(row["technolgies"])
    for t in techs:
        if t["category"] in relevant_cats:
            records.append({
                "group_name": row["group_name"],
                "site_name": row["site_name"],
                "product": t["product"]
            })

tech_df = pd.DataFrame(records).drop_duplicates(subset=["group_name", "site_name", "product"])

# Top 10 groups
top_groups = tech_df.groupby("group_name")["site_name"].nunique().sort_values(ascending=False).head(100).index
filtered = tech_df[tech_df["group_name"].isin(top_groups)]

# Build site-level binary matrix: does site X (of group G) have tech T?
site_group = filtered[["group_name", "site_name"]].drop_duplicates()
site_tech = filtered.groupby("site_name")["product"].apply(set).to_dict()

total_sites = site_group["site_name"].nunique()

# For each group, top 10 techs, compute OR
results = []
for group in top_groups:
    group_sites = set(site_group[site_group["group_name"] == group]["site_name"])
    other_sites = set(site_group[site_group["group_name"] != group]["site_name"])
    
    # Top 10 techs for this group
    group_techs = (
        filtered[filtered["group_name"] == group]
        .groupby("product").size()
        .sort_values(ascending=False).head(100).index
    )
    
    for tech in group_techs:
        a = sum(1 for s in group_sites if tech in site_tech.get(s, set()))   # group + tech
        b = len(group_sites) - a                                              # group + no tech
        c = sum(1 for s in other_sites if tech in site_tech.get(s, set()))   # other + tech
        d = len(other_sites) - c                                              # other + no tech
        
        table = [[a, b], [c, d]]
        odds_ratio, p_value = fisher_exact(table)
        
        results.append({
            "group_name": group,
            "product": tech,
            "a_group_has": a,
            "b_group_no": b,
            "c_other_has": c,
            "d_other_no": d,
            "odds_ratio": round(odds_ratio, 3),
            "p_value": p_value
        })

result_df = pd.DataFrame(results)
result_df = (
    result_df.sort_values("odds_ratio", ascending=False)
    .groupby("group_name", sort=False)
    .head(10)
    .sort_values(["group_name", "odds_ratio"], ascending=[True, False])
)
#
# reversing this pd.set_option('display.float_format', lambda x: f'{x:.2e}')


# save in a file
result_df["p_value"] = result_df["p_value"].apply(lambda x: f'{x:.2e}')
result_df["odds_ratio"] = result_df["odds_ratio"].apply(lambda x: round(x, 3))

result_df.to_csv("output/tech_odds_ratios.csv", index=False)
print(result_df.to_string(index=False))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# ── 1. Caricamento e pulizia ──────────────────────────────────────────
df = pd.read_csv("data/swdb_prodcat.csv")

spending_cols = ["site_hw_spending", "site_ict_spending", "site_sw_spending", "site_it_services_spending"]
df = df.dropna(subset=["group_name"] + spending_cols)
for col in spending_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
df = df.dropna(subset=spending_cols)
df = df[(df[spending_cols] > 0).all(axis=1)].copy()

# totale IT spending
df["total_it_spending"] = df[spending_cols].sum(axis=1)

# spending per dipendente (intensità)
df["it_per_employee"] = np.where(
    df["site_employees"] > 0,
    df["total_it_spending"] / df["site_employees"],
    np.nan
)
# spending come % revenue
df["it_pct_revenue"] = np.where(
    df["site_revenue_usd"] > 0,
    df["total_it_spending"] / df["site_revenue_usd"] * 100,
    np.nan
)

print(f"Record totali: {len(df)}")
print(f"Gruppi unici: {df['group_name'].nunique()}")
print()

# statistiche generali spending
for col in spending_cols + ["total_it_spending"]:
    print(f"{col}: median=${df[col].median():,.0f}, mean=${df[col].mean():,.0f}")
print(f"IT per employee: median=${df['it_per_employee'].median():,.0f}")
print(f"IT as % revenue: median={df['it_pct_revenue'].median():.2f}%")
print()

# ── 2. Top 10 gruppi ─────────────────────────────────────────────────
MIN_VICTIMS = 10
group_counts = df["group_name"].value_counts()
active_groups = group_counts[group_counts >= MIN_VICTIMS].index
df_active = df[df["group_name"].isin(active_groups)].copy()
top10 = group_counts.head(10).index.tolist()

# ── 3. Statistiche per gruppo: ogni tipo di spending ──────────────────
metrics = {
    "total_it_spending": "Total IT Spending",
    "site_ict_spending": "ICT Spending",
    "site_sw_spending": "Software Spending",
    "site_hw_spending": "Hardware Spending",
    "site_it_services_spending": "IT Services Spending",
    "it_per_employee": "IT Spending per Employee",
    "it_pct_revenue": "IT Spending as % Revenue"
}

all_stats = {}
for metric, label in metrics.items():
    stats_df = (
        df_active.groupby("group_name")[metric]
        .agg(["count", "mean", "median", "std"])
        .rename(columns={"count": "n_victims"})
        .sort_values("median", ascending=False)
    )
    all_stats[metric] = stats_df
    
    print("═" * 85)
    print(f"{label.upper()} PER GRUPPO (ordinate per mediana)")
    print("═" * 85)
    display = stats_df.copy()
    if "pct" not in metric:
        for c in ["mean", "median", "std"]:
            display[c] = display[c].apply(lambda x: f"${x:,.0f}" if pd.notna(x) else "N/A")
    else:
        for c in ["mean", "median", "std"]:
            display[c] = display[c].apply(lambda x: f"{x:.2f}%" if pd.notna(x) else "N/A")
    print(display.to_string())
    print()

# ── 4. Kruskal-Wallis per ogni metrica ───────────────────────────────
print("═" * 85)
print("KRUSKAL-WALLIS TESTS")
print("═" * 85)
for metric, label in metrics.items():
    groups_data = [g[metric].dropna().values for _, g in df_active.groupby("group_name")]
    groups_data = [g for g in groups_data if len(g) >= 3]
    if len(groups_data) >= 2:
        h_stat, kw_p = stats.kruskal(*groups_data)
        sig = "✓ Significativo" if kw_p < 0.05 else "✗ Non significativo"
        print(f"  {label:<35} H={h_stat:>8.2f}  p={kw_p:.4e}  {sig}")
print()

# ── 5. Classificazione spending ───────────────────────────────────────
# classi basate su quartili del totale IT spending
q_labels = ["Q1 (Low)", "Q2", "Q3", "Q4 (High)"]
df_active["spending_quartile"] = pd.qcut(df_active["total_it_spending"], 4, labels=q_labels)

spend_dist = (
    df_active[df_active["group_name"].isin(top10)]
    .groupby(["group_name", "spending_quartile"], observed=True)
    .size()
    .unstack(fill_value=0)
)
spend_pct = spend_dist.div(spend_dist.sum(axis=1), axis=0) * 100

print("═" * 85)
print("% VITTIME PER QUARTILE DI IT SPENDING – TOP 10 GRUPPI")
print("═" * 85)
print(spend_pct.round(1).to_string())
print()

# chi-squared
chi2, p_val, dof, _ = stats.chi2_contingency(spend_dist)
print(f"Chi-squared: χ²={chi2:.2f}, df={dof}, p={p_val:.4e}")
print()

# ── 6. Composizione spending: dove investono le vittime ───────────────
# % di ogni componente sul totale
for col, short in [("site_hw_spending", "hw_pct"), ("site_ict_spending", "ict_pct"),
                    ("site_sw_spending", "sw_pct"), ("site_it_services_spending", "services_pct")]:
    df_active[short] = df_active[col] / df_active["total_it_spending"] * 100

comp_stats = (
    df_active[df_active["group_name"].isin(top10)]
    .groupby("group_name")[["hw_pct", "ict_pct", "sw_pct", "services_pct"]]
    .median()
    .sort_values("sw_pct", ascending=False)
)
comp_stats.columns = ["HW %", "ICT %", "SW %", "IT Services %"]
print("═" * 85)
print("COMPOSIZIONE MEDIANA SPENDING (% del totale IT) – TOP 10 GRUPPI")
print("═" * 85)
print(comp_stats.round(1).to_string())
print()

# ── 7. Visualizzazioni ──────────────────────────────────────────────
fig, axes = plt.subplots(2, 3, figsize=(24, 14))
fig.suptitle("Ransomware Groups: Victim IT Spending Analysis", fontsize=16, fontweight="bold")

df_top10 = df_active[df_active["group_name"].isin(top10)]

# 7a. Boxplot Total IT Spending
ax = axes[0, 0]
order = df_top10.groupby("group_name")["total_it_spending"].median().sort_values(ascending=False).index
sns.boxplot(data=df_top10, y="group_name", x="total_it_spending", order=order, ax=ax, fliersize=2)
ax.set_xscale("log")
ax.set_xlabel("Total IT Spending (log)")
ax.set_ylabel("")
ax.set_title("Total IT Spending – Top 10 Groups")
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${x/1e6:.0f}M" if x >= 1e6 else f"${x/1e3:.0f}K"))

# 7b. Boxplot Software Spending
ax = axes[0, 1]
order_sw = df_top10.groupby("group_name")["site_sw_spending"].median().sort_values(ascending=False).index
sns.boxplot(data=df_top10, y="group_name", x="site_sw_spending", order=order_sw, ax=ax, fliersize=2)
ax.set_xscale("log")
ax.set_xlabel("Software Spending (log)")
ax.set_ylabel("")
ax.set_title("Software Spending – Top 10 Groups")
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${x/1e6:.0f}M" if x >= 1e6 else f"${x/1e3:.0f}K"))

# 7c. Boxplot ICT Spending
ax = axes[0, 2]
order_ict = df_top10.groupby("group_name")["site_ict_spending"].median().sort_values(ascending=False).index
sns.boxplot(data=df_top10, y="group_name", x="site_ict_spending", order=order_ict, ax=ax, fliersize=2)
ax.set_xscale("log")
ax.set_xlabel("ICT Spending (log)")
ax.set_ylabel("")
ax.set_title("ICT Spending – Top 10 Groups")
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${x/1e6:.0f}M" if x >= 1e6 else f"${x/1e3:.0f}K"))

# 7d. Heatmap quartili spending
ax = axes[1, 0]
spend_pct_sorted = spend_pct.loc[
    spend_pct.index.to_series()
    .map(lambda x: all_stats["total_it_spending"].loc[x, "median"])
    .sort_values(ascending=False)
    .index
]
sns.heatmap(spend_pct_sorted, annot=True, fmt=".0f", cmap="YlOrRd", ax=ax, cbar_kws={"label": "%"})
ax.set_title("% Victims by IT Spending Quartile")
ax.set_ylabel("")

# 7e. Stacked bar: composizione spending
ax = axes[1, 1]
comp_plot = comp_stats.loc[comp_stats.index.isin(top10)]
comp_plot[["HW %", "ICT %", "SW %", "IT Services %"]].plot(
    kind="barh", stacked=True, ax=ax,
    color=["#4472C4", "#ED7D31", "#70AD47", "#FFC000"]
)
ax.set_xlabel("% of Total IT Spending")
ax.set_title("Spending Composition by Group")
ax.legend(loc="lower right", fontsize=8)

# 7f. Scatter: IT spending mediana vs n vittime
ax = axes[1, 2]
stat = all_stats["total_it_spending"]
ax.scatter(stat["n_victims"], stat["median"], alpha=0.6, edgecolors="black", linewidth=0.5)
for name, row in stat.iterrows():
    if row["n_victims"] > stat["n_victims"].quantile(0.85) or \
       row["median"] > stat["median"].quantile(0.9) or \
       row["median"] < stat["median"].quantile(0.1):
        ax.annotate(name, (row["n_victims"], row["median"]), fontsize=7, alpha=0.8)
ax.set_xlabel("Number of Victims")
ax.set_ylabel("Median Total IT Spending")
ax.set_yscale("log")
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${x/1e6:.0f}M" if x >= 1e6 else f"${x/1e3:.0f}K"))
ax.set_title("Volume vs IT Spending Preference")

plt.tight_layout()
plt.savefig("ransomware_spending_analysis.png", dpi=150, bbox_inches="tight")
plt.show()
print("\n✓ Grafico salvato: ransomware_spending_analysis.png")

# ── 8. Correlazione spending vs attacked ──────────────────────────────
print("═" * 85)
print("CORRELAZIONE: chi spende MENO in IT viene attaccato di più?")
print("═" * 85)
# per ogni gruppo, calcola la correlazione tra spending e frequenza di attacco
# confronta mediana spending vittime vs mediana spending generale
overall_median = df["total_it_spending"].median()
overall_sw_median = df["site_sw_spending"].median()
overall_ict_median = df["site_ict_spending"].median()

print(f"\nMediana generale: Total IT=${overall_median:,.0f}, SW=${overall_sw_median:,.0f}, ICT=${overall_ict_median:,.0f}")
print()
print(f"{'Group':<20} {'Total IT':>12} {'vs Overall':>12} {'SW':>12} {'vs Overall':>12} {'ICT':>12} {'vs Overall':>12}")
print("-" * 95)
for group in top10:
    g = df_active[df_active["group_name"] == group]
    med_it = g["total_it_spending"].median()
    med_sw = g["site_sw_spending"].median()
    med_ict = g["site_ict_spending"].median()
    
    diff_it = ((med_it - overall_median) / overall_median) * 100
    diff_sw = ((med_sw - overall_sw_median) / overall_sw_median) * 100
    diff_ict = ((med_ict - overall_ict_median) / overall_ict_median) * 100
    
    print(f"{group:<20} ${med_it:>10,.0f} {diff_it:>+10.1f}% ${med_sw:>10,.0f} {diff_sw:>+10.1f}% ${med_ict:>10,.0f} {diff_ict:>+10.1f}%")

# ── 9. Export ─────────────────────────────────────────────────────────
export = all_stats["total_it_spending"].copy()
export["median_sw"] = all_stats["site_sw_spending"]["median"]
export["median_ict"] = all_stats["site_ict_spending"]["median"]
export["median_hw"] = all_stats["site_hw_spending"]["median"]
export["median_services"] = all_stats["site_it_services_spending"]["median"]
export.to_csv("group_spending_profile.csv")
print("\n✓ Esportato: group_spending_profile.csv")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# ── 1. Caricamento e pulizia ──────────────────────────────────────────
df = pd.read_csv("data/swdb_tech.csv")
df = df.dropna(subset=["site_employees", "group_name"])
df = df[df["site_employees"] > 0].copy()

print(f"Record totali dopo pulizia: {len(df)}")
print(f"Gruppi unici: {df['group_name'].nunique()}")
print(f"Range dipendenti: {df['site_employees'].min()} – {df['site_employees'].max()}")
print(f"Mediana dipendenti: {df['site_employees'].median()}")
print()

# ── 2. Statistiche descrittive per gruppo ─────────────────────────────
MIN_VICTIMS = 10
group_counts = df["group_name"].value_counts()
active_groups = group_counts[group_counts >= MIN_VICTIMS].index
df_active = df[df["group_name"].isin(active_groups)].copy()

print(f"Gruppi con >= {MIN_VICTIMS} vittime: {len(active_groups)}")
print()

group_stats = (
    df_active.groupby("group_name")["site_employees"]
    .agg(["count", "mean", "median", "std", "min", "max"])
    .rename(columns={"count": "n_victims"})
    .sort_values("median", ascending=False)
)
print("═" * 80)
print("STATISTICHE DIPENDENTI PER GRUPPO (ordinate per mediana)")
print("═" * 80)
print(group_stats.round(1).to_string())
print()

# ── 3. Classificazione dimensione azienda ─────────────────────────────
bins = [0, 10, 50, 250, 1000, np.inf]
labels = ["Micro (1-10)", "Small (11-50)", "Medium (51-250)", "Large (251-1000)", "Enterprise (1000+)"]
df_active["size_class"] = pd.cut(df_active["site_employees"], bins=bins, labels=labels)

size_dist = (
    df_active.groupby(["group_name", "size_class"], observed=True)
    .size()
    .unstack(fill_value=0)
)
size_pct = size_dist.div(size_dist.sum(axis=1), axis=0) * 100

print("═" * 80)
print("DISTRIBUZIONE % DIMENSIONE VITTIME PER GRUPPO")
print("═" * 80)
print(size_pct.round(1).to_string())
print()

# ── 4. Chi-squared test ──────────────────────────────────────────────
chi2, p_val, dof, expected = stats.chi2_contingency(size_dist)
print(f"Chi-squared test (gruppo vs dimensione azienda):")
print(f"  χ² = {chi2:.2f}, df = {dof}, p-value = {p_val:.4e}")
print(f"  → {'Differenza significativa' if p_val < 0.05 else 'Nessuna differenza significativa'} (α=0.05)")
print()

# ── 5. Kruskal-Wallis test ───────────────────────────────────────────
groups_data = [g["site_employees"].values for _, g in df_active.groupby("group_name")]
h_stat, kw_p = stats.kruskal(*groups_data)
print(f"Kruskal-Wallis test (dipendenti ~ gruppo):")
print(f"  H = {h_stat:.2f}, p-value = {kw_p:.4e}")
print(f"  → {'Differenza significativa' if kw_p < 0.05 else 'Nessuna differenza significativa'} (α=0.05)")
print()

# ── 6. Top gruppi che preferiscono aziende grandi vs piccole ─────────
print("═" * 80)
print("TOP 10 GRUPPI – MEDIANA DIPENDENTI PIÙ ALTA (preferiscono aziende grandi)")
print("═" * 80)
top_large = group_stats.head(10)[["n_victims", "median", "mean"]]
print(top_large.round(1).to_string())
print()

print("═" * 80)
print("TOP 10 GRUPPI – MEDIANA DIPENDENTI PIÙ BASSA (preferiscono aziende piccole)")
print("═" * 80)
top_small = group_stats.tail(10)[["n_victims", "median", "mean"]]
print(top_small.round(1).to_string())
print()

# ── 7. Visualizzazioni ──────────────────────────────────────────────
fig, axes = plt.subplots(2, 2, figsize=(18, 14))
fig.suptitle("Ransomware Groups: Victim Selection by Company Size (Employees)", fontsize=16, fontweight="bold")

# 7a. Boxplot top 15 gruppi per numero vittime
top15 = group_counts.head(15).index
df_top15 = df_active[df_active["group_name"].isin(top15)]
order = df_top15.groupby("group_name")["site_employees"].median().sort_values(ascending=False).index

ax = axes[0, 0]
sns.boxplot(data=df_top15, y="group_name", x="site_employees", order=order, ax=ax, fliersize=2)
ax.set_xscale("log")
ax.set_xlabel("Employees (log scale)")
ax.set_ylabel("")
ax.set_title("Employee Distribution – Top 15 Groups by Volume")

# 7b. Heatmap distribuzione percentuale
top20_groups = group_counts.head(20).index
size_pct_top20 = size_pct.loc[size_pct.index.isin(top20_groups)]
sort_order = (
    size_pct_top20.index
    .to_series()
    .map(lambda x: group_stats.loc[x, "median"])
    .sort_values(ascending=False)
    .index
)
size_pct_top20 = size_pct_top20.loc[sort_order]

ax = axes[0, 1]
sns.heatmap(size_pct_top20, annot=True, fmt=".0f", cmap="YlOrRd", ax=ax, cbar_kws={"label": "%"})
ax.set_title("% Victims by Size Class – Top 20 Groups")
ax.set_ylabel("")

# 7c. Violin plot top 10 gruppi
top10 = group_counts.head(10).index
df_top10 = df_active[df_active["group_name"].isin(top10)]

ax = axes[1, 0]
sns.violinplot(data=df_top10, y="group_name", x="site_employees",
               order=df_top10.groupby("group_name")["site_employees"].median().sort_values(ascending=False).index,
               ax=ax, inner="quartile", cut=0)
ax.set_xscale("log")
ax.set_xlabel("Employees (log scale)")
ax.set_ylabel("")
ax.set_title("Employee Distribution (Violin) – Top 10 Groups")

# 7d. Scatter: mediana dipendenti vs numero vittime
ax = axes[1, 1]
ax.scatter(group_stats["n_victims"], group_stats["median"], alpha=0.6, edgecolors="black", linewidth=0.5)
for name, row in group_stats.iterrows():
    if row["n_victims"] > group_stats["n_victims"].quantile(0.85) or row["median"] > group_stats["median"].quantile(0.9):
        ax.annotate(name, (row["n_victims"], row["median"]), fontsize=7, alpha=0.8)
ax.set_xlabel("Number of Victims")
ax.set_ylabel("Median Employees")
ax.set_yscale("log")
ax.set_title("Volume vs Target Size Preference")

plt.tight_layout()
plt.savefig("ransomware_employee_analysis.png", dpi=150, bbox_inches="tight")
plt.show()
print("\n✓ Grafico salvato: ransomware_employee_analysis.png")

# ── 8. Export summary CSV ────────────────────────────────────────────
export = group_stats.copy()
export["pct_micro"] = size_pct.get("Micro (1-10)", 0)
export["pct_small"] = size_pct.get("Small (11-50)", 0)
export["pct_medium"] = size_pct.get("Medium (51-250)", 0)
export["pct_large"] = size_pct.get("Large (251-1000)", 0)
export["pct_enterprise"] = size_pct.get("Enterprise (1000+)", 0)
export.to_csv("group_employee_profile.csv")
print("✓ Summary esportato: group_employee_profile.csv")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from collections import Counter

# ── 1. Caricamento e pulizia ──────────────────────────────────────────
df = pd.read_csv("data/swdb_tech.csv")
df = df.dropna(subset=["technologies", "group_name"])
df = df[df["technologies"].str.strip() != ""].copy()

print(f"Record totali con tecnologie: {len(df)}")
print(f"Gruppi unici: {df['group_name'].nunique()}")

# esplodi le tecnologie
df["tech_list"] = df["technologies"].str.split("; ")
df["tech_count"] = df["tech_list"].apply(len)

print(f"Mediana tecnologie per sito: {df['tech_count'].median():.0f}")
print(f"Range: {df['tech_count'].min()} – {df['tech_count'].max()}")
print()

# ── 2. Tecnologie più comuni overall ─────────────────────────────────
all_techs = df["tech_list"].explode()
tech_freq = all_techs.value_counts()
n_sites = len(df)

print("═" * 80)
print(f"TOP 30 TECNOLOGIE PIÙ COMUNI (su {n_sites} siti vittima)")
print("═" * 80)
for tech, count in tech_freq.head(30).items():
    print(f"  {tech:<45} {count:>5}  ({count/n_sites*100:.1f}%)")
print()

# ── 3. Statistiche tech_count per gruppo ──────────────────────────────
MIN_VICTIMS = 10
group_counts = df["group_name"].value_counts()
active_groups = group_counts[group_counts >= MIN_VICTIMS].index
df_active = df[df["group_name"].isin(active_groups)].copy()

tech_count_stats = (
    df_active.groupby("group_name")["tech_count"]
    .agg(["count", "mean", "median", "std"])
    .rename(columns={"count": "n_victims"})
    .sort_values("median", ascending=False)
)
print("═" * 80)
print("NUMERO TECNOLOGIE PER SITO – PER GRUPPO (proxy di complessità stack)")
print("═" * 80)
print(tech_count_stats.round(1).to_string())
print()

# Kruskal-Wallis su tech_count
kw_groups = [g["tech_count"].values for _, g in df_active.groupby("group_name")]
h_stat, kw_p = stats.kruskal(*kw_groups)
print(f"Kruskal-Wallis (tech_count ~ gruppo): H={h_stat:.2f}, p={kw_p:.4e}")
print(f"  → {'Significativo' if kw_p < 0.05 else 'Non significativo'}")
print()

# ── 4. Tecnologie distintive per gruppo (odds ratio) ─────────────────
# per ogni gruppo, calcola quali tech sono sovra/sotto-rappresentate
# rispetto al resto del dataset

print("═" * 80)
print("TECNOLOGIE DISTINTIVE PER GRUPPO (Odds Ratio vs resto del dataset)")
print("═" * 80)

# crea matrice binaria sito x tecnologia (solo top 100 tech per efficienza)
top_techs = tech_freq.head(100).index.tolist()

# esplodi e filtra
exploded = df_active[["group_name"]].copy()
exploded["technology"] = df_active["tech_list"]
exploded = exploded.explode("technology")
exploded = exploded[exploded["technology"].isin(top_techs)]

# calcola odds ratio per i top 15 gruppi
top15_groups = group_counts.head(15).index
or_results = []

for group in top15_groups:
    group_mask = exploded["group_name"] == group
    group_techs = exploded[group_mask]["technology"].value_counts()
    other_techs = exploded[~group_mask]["technology"].value_counts()
    
    n_group = df_active[df_active["group_name"] == group].shape[0]
    n_other = df_active[df_active["group_name"] != group].shape[0]
    
    for tech in top_techs:
        a = group_techs.get(tech, 0)       # gruppo ha tech
        b = n_group - a                      # gruppo non ha tech
        c = other_techs.get(tech, 0)         # altri hanno tech
        d = n_other - c                      # altri non hanno tech
        
        # aggiungi 0.5 per evitare divisione per zero
        odds_ratio = ((a + 0.5) * (d + 0.5)) / ((b + 0.5) * (c + 0.5))
        
        # fisher exact per significatività (solo se abbastanza osservazioni)
        if a >= 3:
            _, p_fisher = stats.fisher_exact([[a, b], [c, d]])
        else:
            p_fisher = 1.0
        
        or_results.append({
            "group": group,
            "technology": tech,
            "group_count": a,
            "group_pct": a / n_group * 100,
            "other_pct": c / n_other * 100,
            "odds_ratio": odds_ratio,
            "p_value": p_fisher
        })

or_df = pd.DataFrame(or_results)

# mostra top 5 tech sovra-rappresentate per ogni gruppo
for group in top15_groups:
    g = or_df[(or_df["group"] == group) & (or_df["p_value"] < 0.05)]
    over = g.nlargest(5, "odds_ratio")
    under = g.nsmallest(5, "odds_ratio")
    
    print(f"\n▶ {group.upper()} ({group_counts[group]} vittime)")
    if len(over) > 0:
        print("  SOVRA-rappresentate:")
        for _, r in over.iterrows():
            print(f"    {r['technology']:<40} OR={r['odds_ratio']:.2f}  ({r['group_pct']:.0f}% vs {r['other_pct']:.0f}%)  p={r['p_value']:.4f}")
    if len(under) > 0:
        print("  SOTTO-rappresentate:")
        for _, r in under.iterrows():
            print(f"    {r['technology']:<40} OR={r['odds_ratio']:.2f}  ({r['group_pct']:.0f}% vs {r['other_pct']:.0f}%)  p={r['p_value']:.4f}")

print()

# ── 5. Categorizzazione tecnologie ───────────────────────────────────
tech_categories = {
    "CMS": ["WordPress", "Drupal", "Joomla", "Magento Commerce", "Shopify", "Wix", "Squarespace", "PrestaShop"],
    "CDN": ["Cloudflare CDN", "Akamai CDN", "Amazon CloudFront", "Fastly", "CloudFlare"],
    "Analytics": ["Google Analytics", "Google Tag Manager", "Hotjar", "Adobe Tag Management", "Facebook Pixel"],
    "JavaScript": ["jQuery", "jQuery Migrate", "jQuery UI", "React", "Angular", "Vue.js", "Bootstrap", "Modernizr"],
    "Server": ["nginx", "Apache", "Microsoft IIS", "LiteSpeed", "Web Server"],
    "Language": ["PHP", "ASP.NET", "Python", "Java", "Ruby"],
    "Security": ["reCAPTCHA", "SSL by Default", "Thawte Trusted Site Seal", "Let's Encrypt"],
    "Email": ["Microsoft Office", "Google Workspace", "Amazon Simple Email Service (SES)", "Mailchimp"],
    "Payment": ["Stripe", "PayPal", "VISA", "Mastercard", "American Express", "Authorize.net", "Braintree", "Payments"],
    "CRM": ["Salesforce", "HubSpot", "SAP CRM", "Pipeliner CRM", "Zendesk"],
    "Cloud": ["Amazon Web Services", "Microsoft Azure", "Google Cloud"],
    "DNS": ["Cloudflare DNS", "Amazon Route 53", "GoDaddy DNS"],
}

# conta categorie per sito
def get_categories(tech_list):
    cats = set()
    for tech in tech_list:
        for cat, members in tech_categories.items():
            if tech in members:
                cats.add(cat)
    return list(cats)

df_active["tech_categories"] = df_active["tech_list"].apply(get_categories)
df_active["n_categories"] = df_active["tech_categories"].apply(len)

# esplodi categorie
cat_exploded = df_active[["group_name"]].copy()
cat_exploded["category"] = df_active["tech_categories"]
cat_exploded = cat_exploded.explode("category")
cat_exploded = cat_exploded.dropna()

# matrice gruppo x categoria (%)
cat_matrix = (
    cat_exploded.groupby(["group_name", "category"]).size()
    .unstack(fill_value=0)
)
cat_pct = cat_matrix.div(df_active.groupby("group_name").size(), axis=0) * 100
cat_pct = cat_pct.loc[cat_pct.index.isin(top15_groups)]

print("═" * 80)
print("% SITI VITTIMA CON OGNI CATEGORIA TECH – TOP 15 GRUPPI")
print("═" * 80)
print(cat_pct.round(1).to_string())
print()

# ── 6. Visualizzazioni ──────────────────────────────────────────────
fig, axes = plt.subplots(2, 2, figsize=(20, 16))
fig.suptitle("Ransomware Groups: Victim Technology Stack Analysis", fontsize=16, fontweight="bold")

# 6a. Top 25 tecnologie più comuni
ax = axes[0, 0]
top25 = tech_freq.head(25)
bars = ax.barh(range(len(top25)), top25.values, color=sns.color_palette("viridis", len(top25)))
ax.set_yticks(range(len(top25)))
ax.set_yticklabels(top25.index, fontsize=8)
ax.set_xlabel("Number of Victim Sites")
ax.set_title("Top 25 Most Common Technologies Among Victims")
ax.invert_yaxis()

# 6b. Heatmap categorie per gruppo
ax = axes[0, 1]
# ordina per mediana tech_count
cat_pct_sorted = cat_pct.loc[
    cat_pct.index.to_series()
    .map(lambda x: tech_count_stats.loc[x, "median"] if x in tech_count_stats.index else 0)
    .sort_values(ascending=False)
    .index
]
sns.heatmap(cat_pct_sorted, annot=True, fmt=".0f", cmap="YlOrRd", ax=ax, 
            cbar_kws={"label": "% of victims"}, linewidths=0.5)
ax.set_title("% Victims with Technology Category – Top 15 Groups")
ax.set_ylabel("")

# 6c. Boxplot tech_count per gruppo (top 15)
df_top15 = df_active[df_active["group_name"].isin(top15_groups)]
order = df_top15.groupby("group_name")["tech_count"].median().sort_values(ascending=False).index

ax = axes[1, 0]
sns.boxplot(data=df_top15, y="group_name", x="tech_count", order=order, ax=ax, fliersize=2)
ax.set_xlabel("Number of Technologies per Site")
ax.set_ylabel("")
ax.set_title("Technology Stack Complexity – Top 15 Groups")

# 6d. Scatter: tech_count mediana vs numero vittime
ax = axes[1, 1]
ax.scatter(tech_count_stats["n_victims"], tech_count_stats["median"], 
           alpha=0.6, edgecolors="black", linewidth=0.5)
for name, row in tech_count_stats.iterrows():
    if row["n_victims"] > tech_count_stats["n_victims"].quantile(0.85) or \
       row["median"] > tech_count_stats["median"].quantile(0.9) or \
       row["median"] < tech_count_stats["median"].quantile(0.1):
        ax.annotate(name, (row["n_victims"], row["median"]), fontsize=7, alpha=0.8)
ax.set_xlabel("Number of Victims")
ax.set_ylabel("Median Technologies per Site")
ax.set_title("Volume vs Stack Complexity Preference")

plt.tight_layout()
plt.savefig("ransomware_technology_analysis.png", dpi=150, bbox_inches="tight")
plt.show()
print("\n✓ Grafico salvato: ransomware_technology_analysis.png")

# ── 7. Export ─────────────────────────────────────────────────────────
or_df.to_csv("group_tech_odds_ratios.csv", index=False)
cat_pct.to_csv("group_tech_categories.csv")
print("✓ Esportati: group_tech_odds_ratios.csv, group_tech_categories.csv")
```

```python
import pandas as pd
import numpy as np
from scipy import stats
from openpyxl import Workbook
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side
from openpyxl.utils import get_column_letter

# ── 1. Caricamento e pulizia ──────────────────────────────────────────
df = pd.read_csv("data/swdb_tech.csv")
df = df.dropna(subset=["technologies", "group_name"])
df = df[df["technologies"].str.strip() != ""].copy()

df["tech_list"] = df["technologies"].str.split("; ")
df["tech_count"] = df["tech_list"].apply(len)

# esplodi e filtra tech con < 10 occorrenze
all_techs = df["tech_list"].explode()
tech_freq = all_techs.value_counts()
valid_techs = tech_freq[tech_freq >= 10].index.tolist()
print(f"Tecnologie totali: {tech_freq.shape[0]}")
print(f"Tecnologie con >= 10 occorrenze: {len(valid_techs)}")

# top 10 gruppi per numero vittime
group_counts = df["group_name"].value_counts()
top10_groups = group_counts.head(10).index.tolist()
df_active = df[df["group_name"].isin(top10_groups)].copy()

print(f"Top 10 gruppi: {top10_groups}")
print()

# ── 2. Tabella summary: tech stack complexity ─────────────────────────
summary = (
    df_active.groupby("group_name")["tech_count"]
    .agg(["count", "mean", "median", "std", "min", "max"])
    .rename(columns={"count": "n_victims"})
    .sort_values("mean", ascending=False)
)
summary = summary.round(2)
print("TECH STACK COMPLEXITY – TOP 10 GRUPPI")
print(summary.to_string())
print()

# ── 3. Odds Ratio per ogni gruppo ─────────────────────────────────────
exploded = df_active[["group_name"]].copy()
exploded["technology"] = df_active["tech_list"]
exploded = exploded.explode("technology")
exploded = exploded[exploded["technology"].isin(valid_techs)]

or_tables = {}

for group in top10_groups:
    group_mask = exploded["group_name"] == group
    group_techs = exploded[group_mask]["technology"].value_counts()
    other_techs = exploded[~group_mask]["technology"].value_counts()

    n_group = df_active[df_active["group_name"] == group].shape[0]
    n_other = df_active[df_active["group_name"] != group].shape[0]

    rows = []
    for tech in valid_techs:
        a = group_techs.get(tech, 0)
        b = n_group - a
        c = other_techs.get(tech, 0)
        d = n_other - c

        odds_ratio = ((a + 0.5) * (d + 0.5)) / ((b + 0.5) * (c + 0.5))

        if a >= 3:
            _, p_fisher = stats.fisher_exact([[a, b], [c, d]])
        else:
            p_fisher = 1.0

        rows.append({
            "technology": tech,
            "group_count": int(a),
            "group_pct": round(a / n_group * 100, 1),
            "others_count": int(c),
            "others_pct": round(c / n_other * 100, 1),
            "odds_ratio": round(odds_ratio, 3),
            "p_value": p_fisher,
            "direction": "OVER" if odds_ratio > 1 else "UNDER"
        })

    or_df = pd.DataFrame(rows)
    # solo significativi, top 10 per OR (sovra-rappresentate)
    sig = or_df[or_df["p_value"] < 0.05].copy()
    top_or = sig.nlargest(10, "odds_ratio")
    or_tables[group] = top_or
    
    print(f"▶ {group.upper()} – Top 10 significant OR")
    print(top_or[["technology", "group_pct", "others_pct", "odds_ratio", "p_value"]].to_string(index=False))
    print()

# ── 4. Creazione Excel ───────────────────────────────────────────────
wb = Workbook()

# stili
header_font = Font(name="Arial", bold=True, color="FFFFFF", size=11)
header_fill = PatternFill("solid", fgColor="2F5496")
subheader_fill = PatternFill("solid", fgColor="D6E4F0")
subheader_font = Font(name="Arial", bold=True, size=10)
data_font = Font(name="Arial", size=10)
border = Border(
    bottom=Side(style="thin", color="CCCCCC"),
    right=Side(style="thin", color="CCCCCC")
)
pct_format = '0.0"%"'
or_format = '0.000'
sci_format = '0.00E+00'

def style_header(ws, row, max_col):
    for col in range(1, max_col + 1):
        cell = ws.cell(row=row, column=col)
        cell.font = header_font
        cell.fill = header_fill
        cell.alignment = Alignment(horizontal="center", wrap_text=True)

def style_data_row(ws, row, max_col, alt=False):
    fill = PatternFill("solid", fgColor="F2F2F2") if alt else PatternFill("solid", fgColor="FFFFFF")
    for col in range(1, max_col + 1):
        cell = ws.cell(row=row, column=col)
        cell.font = data_font
        cell.fill = fill
        cell.border = border
        cell.alignment = Alignment(horizontal="center") if col > 1 else Alignment(horizontal="left")

# ── Sheet 1: Summary ──────────────────────────────────────────────────
ws = wb.active
ws.title = "Stack Complexity"
ws.sheet_properties.tabColor = "2F5496"

ws.cell(row=1, column=1, value="Technology Stack Complexity – Top 10 Ransomware Groups by Victim Count")
ws.cell(row=1, column=1).font = Font(name="Arial", bold=True, size=13, color="2F5496")
ws.merge_cells(start_row=1, start_column=1, end_row=1, end_column=7)

headers = ["Group", "N Victims", "Mean Techs", "Median Techs", "Std Dev", "Min", "Max"]
for col, h in enumerate(headers, 1):
    ws.cell(row=3, column=col, value=h)
style_header(ws, 3, len(headers))

for i, (group, row) in enumerate(summary.iterrows()):
    r = 4 + i
    ws.cell(row=r, column=1, value=group)
    ws.cell(row=r, column=2, value=int(row["n_victims"]))
    ws.cell(row=r, column=3, value=row["mean"])
    ws.cell(row=r, column=4, value=row["median"])
    ws.cell(row=r, column=5, value=row["std"])
    ws.cell(row=r, column=6, value=int(row["min"]))
    ws.cell(row=r, column=7, value=int(row["max"]))
    style_data_row(ws, r, len(headers), alt=(i % 2 == 1))

ws.column_dimensions["A"].width = 20
for col in range(2, 8):
    ws.column_dimensions[get_column_letter(col)].width = 14

# ── Sheets 2-11: Odds Ratio per gruppo ────────────────────────────────
# colore tab per gruppo
tab_colors = ["4472C4", "ED7D31", "A5A5A5", "FFC000", "5B9BD5",
              "70AD47", "264478", "9B59B6", "E74C3C", "1ABC9C"]

for idx, group in enumerate(top10_groups):
    sheet_name = group[:28]  # max 31 chars per sheet name
    ws = wb.create_sheet(title=sheet_name)
    ws.sheet_properties.tabColor = tab_colors[idx % len(tab_colors)]

    n_victims = group_counts[group]
    ws.cell(row=1, column=1, value=f"{group.upper()} – Top 10 Significant Odds Ratios (n={n_victims} victims)")
    ws.cell(row=1, column=1).font = Font(name="Arial", bold=True, size=12, color=tab_colors[idx % len(tab_colors)])
    ws.merge_cells(start_row=1, start_column=1, end_row=1, end_column=7)

    headers = ["Technology", "Group Count", "Group %", "Others %", "Odds Ratio", "p-value", "Interpretation"]
    for col, h in enumerate(headers, 1):
        ws.cell(row=3, column=col, value=h)
    style_header(ws, 3, len(headers))

    top_or = or_tables[group]
    for i, (_, row) in enumerate(top_or.iterrows()):
        r = 4 + i
        ws.cell(row=r, column=1, value=row["technology"])
        ws.cell(row=r, column=2, value=row["group_count"])
        ws.cell(row=r, column=3, value=row["group_pct"])
        ws.cell(row=r, column=3).number_format = '0.0'
        ws.cell(row=r, column=4, value=row["others_pct"])
        ws.cell(row=r, column=4).number_format = '0.0'
        ws.cell(row=r, column=5, value=row["odds_ratio"])
        ws.cell(row=r, column=5).number_format = '0.000'
        ws.cell(row=r, column=6, value=row["p_value"])
        ws.cell(row=r, column=6).number_format = '0.00E+00'
        
        # interpretazione leggibile
        diff = row["group_pct"] - row["others_pct"]
        ws.cell(row=r, column=7, value=f"{row['group_pct']:.0f}% vs {row['others_pct']:.0f}% (+{diff:.0f}pp)")
        
        style_data_row(ws, r, len(headers), alt=(i % 2 == 1))
        
        # colora OR alto in rosso/arancione
        or_cell = ws.cell(row=r, column=5)
        if row["odds_ratio"] > 5:
            or_cell.fill = PatternFill("solid", fgColor="FF6B6B")
            or_cell.font = Font(name="Arial", bold=True, size=10, color="FFFFFF")
        elif row["odds_ratio"] > 2:
            or_cell.fill = PatternFill("solid", fgColor="FFB347")
            or_cell.font = Font(name="Arial", bold=True, size=10)

    ws.column_dimensions["A"].width = 40
    ws.column_dimensions["B"].width = 14
    ws.column_dimensions["C"].width = 12
    ws.column_dimensions["D"].width = 12
    ws.column_dimensions["E"].width = 14
    ws.column_dimensions["F"].width = 14
    ws.column_dimensions["G"].width = 25

output_path = "ransomware_tech_odds_ratios.xlsx"
wb.save(output_path)
print(f"\n✓ File salvato: {output_path}")
print(f"  → Sheet 'Stack Complexity': tabella riassuntiva")
print(f"  → 10 sheet con odds ratio per ogni gruppo")
```
